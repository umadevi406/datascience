```python
dfvalid=dfcredit[dfcredit["Class"]==0]
```

```python
dfvalid.shape
```

```python
dfinvalid=dfcredit[dfcredit["Class"]==1]
```

```python
dfinvalid.shape
```

```python
dffinal=pd.concat([dfvalid,dfinvalid])
```

```python
dffinal
```

```python
X=dffinal.drop('Class',axis=1)
```

```python
X.shape
```

```python
y=dffinal["Class"]
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.20)
```

```python
Xtrain.shape,Xtest.shape,X.shape
```

```python
ytrain.shape,ytest.shape,y.shape
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
```

```python
gbc.fit(Xtrain,ytrain)
```

```python
gbc.score(Xtest,ytest)
```

```python
###cross validation
```

```python
from sklearn.model_selection import KFold,cross_val_score
kf=KFold(n_splits=6)
```

```python
gbc1=GradientBoostingClassifier()
```

```python
gbc_score=cross_val_score(gbc1,X,y,cv=kf)
```

```python
gbc_score
```

```python
gbc_score.mean()
```
